```python
import numpy as np
import pandas as pd
import string
from random import sample
```

### Create Database for HW

```python
char = list(string.ascii_letters)
num = list(str(i) for i in range(10))
element = num + char

def sample_password(element):
    return ("".join(sample(element, 6)))
```

```python
passwords = []
password_index = []
names = ["HWMG", "HWTG"]

for name in names:
    if name == "HWMG":
        num = 28
    else:
        num = 36
        
    
    for i in range(1,num):

        password = sample_password(element)
        passwords.append(password)
        code = """CREATE DATABASE %s%02d CHARACTER SET utf8mb4 COLLATE utf8mb4_general_ci;
CREATE USER \'%s%02d\'@\'%%\' IDENTIFIED BY \'%s\'; 
GRANT ALL PRIVILEGES ON %s%02d.* TO \'%s%02d\'@\'%%\'; FLUSH PRIVILEGES; 
GRANT USAGE ON *.* TO \'%s%02d\'@\'%%\'; 
GRANT ALL PRIVILEGES ON `%s%02d`.* TO \'%s%02d\'@\'%%\' WITH GRANT OPTION;""" %(name, i, name, i, password, name, i, name, i, name, i, name, i, name, i)

        password_index.append("%s%02d"%(name, i))
        print(code)
        print()
```

```python
HWG_info = pd.DataFrame({
    "User" : password_index,
    "Password" : passwords
})
```

```python
HWG_info.to_csv("HWG_info.csv", index=False)
```
